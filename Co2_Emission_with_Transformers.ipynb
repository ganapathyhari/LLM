```python
from huggingface_hub import HfApi

api = HfApi()
models = api.list_models(limit=10, cardData=True)
models_list = list(models)

if models_list:
    print("Models with CO2 Emission Data:")
    found_emissions = False
    for model in models_list:
        co2_emissions = model.cardData.get("co2_eq_emissions", "N/A") if model.cardData else "N/A"

        if co2_emissions != "N/A":
            print(f'Model Name: {model.modelId}\nCO2 Emitted during training: {co2_emissions}\n')
            found_emissions = True
    if not found_emissions:
        print("No models with CO2 emission data found within the first 10 models.")
else:
    print("No models found.")
```
